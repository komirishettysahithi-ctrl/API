# Temperature Converter

## SkillCraft Technology Internship

This project converts temperatures between Celsius, Fahrenheit, and Kelvin using Python.

```python
# Temperature Converter

temp = float(input("Enter temperature: "))

from_unit = input("Enter current unit (C/F/K): ").upper()
to_unit = input("Convert to (C/F/K): ").upper()

if from_unit == to_unit:
    result = temp

elif from_unit == "C" and to_unit == "F":
    result = (temp * 9/5) + 32

elif from_unit == "C" and to_unit == "K":
    result = temp + 273.15

elif from_unit == "F" and to_unit == "C":
    result = (temp - 32) * 5/9

elif from_unit == "F" and to_unit == "K":
    result = (temp - 32) * 5/9 + 273.15

elif from_unit == "K" and to_unit == "C":
    result = temp - 273.15

elif from_unit == "K" and to_unit == "F":
    result = (temp - 273.15) * 9/5 + 32

else:
    print("Invalid unit!")
    exit()

print("Converted Temperature:", round(result, 2), to_unit)
```
